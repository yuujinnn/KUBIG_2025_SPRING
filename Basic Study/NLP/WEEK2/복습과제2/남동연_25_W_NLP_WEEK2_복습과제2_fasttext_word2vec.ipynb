## 3. 데이터 전처리

```python
def word_to_jamo(token):
  def to_special_token(jamo): # 경우에 따라 초, 중, 종성이 다 있는 게 아닌 경우도 있다. 이 경우 -를 반환해주는 함수
    if not jamo:
      return '-'
    else:
      return jamo

  decomposed_token = ''
  for char in token:
    try:
      # char(음절)을 초성, 중성, 종성으로 분리
      cho, jung, jong = hgtk.letter.decompose(char)

      # 자모가 빈 문자일 경우 특수문자 -로 대체
      cho = to_special_token(cho)
      jung = to_special_token(jung)
      jong = to_special_token(jong)
      decomposed_token = decomposed_token + cho + jung + jong

    # 만약 char(음절)이 한글이 아닐 경우 자모를 나누지 않고 추가
    except Exception as exception:
      if type(exception).__name__ == 'NotHangulException':
        decomposed_token += char

  # 단어 토큰의 자모 단위 분리 결과를 추가
  return decomposed_token
```

```python
print(word_to_jamo('남동생'))
print(word_to_jamo('야구')) # 야구의 경우 종성이 없으므로 종성 부분을 -로 반환
print(word_to_jamo('남a동12연'))
```

```python
print(mecab.morphs('선물용으로 빨리 받아서 전달했어야 하는 상품이었는데 머그컵만 와서 당황했습니다.'))
```

```python
# mecab으로 형태소를 분리해주고 그 형태소마다 각각 자음모음을 분리해주는 함수
def tokenize_by_jamo(s):
    return [word_to_jamo(token) for token in mecab.morphs(s)]
```

단어를 자모 분리한 것을 역으로 하여 자모 상태를 단어로 다시 결합시키는 함수도 정의해봅시다. 이는 단어의 코사인 유사도를 평가할 때 자모 분리가 된 상태가 아니라 단어 상태로 편리하게 보기 위함입니다.

⭐ **코사인 유사도**: 두 벡터간의 유사도를 측정하는 지표로 두 벡터가 이루는 각도의 코사인 값을 계산하여 얼마나 비슷한지를 측정한다!
- 1: 두 벡터가 동일한 방향 => 완전히 비슷
- 0: 두 벡터가 직각 => 유사도 없음
- -1: 완전히 반대 방향 => 정반대

```python
def jamo_to_word(jamo_sequence):
  tokenized_jamo = []
  index = 0

  # 1. 초기 입력
  # jamo_sequence = 'ㄴㅏㅁㄷㅗㅇㅅㅐㅇ'

  while index < len(jamo_sequence):
    # 문자가 한글(정상적인 자모)이 아닐 경우
    if not hgtk.checker.is_hangul(jamo_sequence[index]):
      tokenized_jamo.append(jamo_sequence[index])
      index = index + 1

    # 문자가 정상적인 자모라면 초성, 중성, 종성을 하나의 토큰으로 간주.
    else:
      tokenized_jamo.append(jamo_sequence[index:index + 3])
      index = index + 3

  # 2. 자모 단위 토큰화 완료
  # tokenized_jamo : ['ㄴㅏㅁ', 'ㄷㅗㅇ', 'ㅅㅐㅇ']

  word = ''
  try:
    for jamo in tokenized_jamo:

      # 초성, 중성, 종성의 묶음으로 추정되는 경우
      if len(jamo) == 3:
        if jamo[2] == "-":
          # 종성이 존재하지 않는 경우
          word = word + hgtk.letter.compose(jamo[0], jamo[1])
        else:
          # 종성이 존재하는 경우
          word = word + hgtk.letter.compose(jamo[0], jamo[1], jamo[2])
      # 한글이 아닌 경우
      else:
        word = word + jamo

  # 복원 중(hgtk.letter.compose) 에러 발생 시 초기 입력 리턴.
  # 복원이 불가능한 경우 예시) 'ㄴ!ㅁㄷㅗㅇㅅㅐㅇ'
  except Exception as exception:
    if type(exception).__name__ == 'NotHangulException':
      return jamo_sequence

  # 3. 단어로 복원 완료
  # word : '남동생'

  return word
```

`get_nearest_neighbors` 함수를 사용하여 '남동생'이라는 단어와 가장 유사도가 높은 단어들(자모 분리된 상태)을 k개만큼 출력해줍니다.

앞서 만든 `jamo_to_word`로 가독성이 좋게 출력해봅시다.

```python
def transform(word_sequence):
  return [(jamo_to_word(word), similarity) for (similarity, word) in word_sequence]
```

이제 word2vec를 사용하여 자모 단위로 분리하는 것이 아닌 단어 단위로 분리하여 임베딩 벡터를 생성해볼 것입니다.

```python
# 간단하게 불용어 정의
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

tokenized_data2 = []
for sentence in tqdm(total_data['reviews'].to_list()):
    tokenized_sentence = mecab.morphs(sentence) # 토큰화
    stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] # 불용어 제거
    tokenized_data2.append(stopwords_removed_sentence)
```

**Your Inference:**

**유사도 측면**:

 먼저 유사도 측면에서 특정 단어의 벡터와 비슷하다고 판단한 단어의 유사도가 더 높게 나타났다


**단어의 의미 측면**: `FastText`의 결과로 나온 유사한 단어가 의미적으로   `Word2Vec`보다 더 가깝다고 볼 수 있다.

-  `남동생`의 경우 `FastText`은 남동생에 들어있는 의미들, "남자", "동생", "아래", "사람", "관계"의 뜻을 더 잘 표현한 단어들이 나타난다. 예를 들어 남친은 남자, 친구는 관계, 남편은 남자+관계, 조카는 남자+아래+관계, 남아= 남자+아래+사람 라고 볼 수 있다.

**단어의 생김새 측면**: `Fasttext`의 결과가 의미적 뿐만 아니라 단어의 생김새 역시 비슷하다.
- `FastText`의 경우 특히 `주문`에서 유사한 단어들로 나온 것들의 생김세가 매우 비슷하다. 예를 들어 주문건,주문서,주문자,주무,주무시와 같이 형태적으로 주문과 매우 비슷한 단어들이 유사한 단어들로 나왔으며 주문할 때 같이 나오는 동사인 "시키다"와 유사한 주무시, 시킨, 시켰었이 나온 걸로 보아 성능이 좋다는 것을 알 수 있다

### **CBOW VS Skip Gram**

 CBOW : ⭐ **여러 단어의 평균으로 학습**
 - 주변 단어들의 정보를 활용하여 풍부한 의미를 포착할 수 있음
 -  희귀 단어의 경우 주변 단어가 적어 정확한 예측 어려움

 => 주로 데이터의 양이 충분하지 않고 문맥 정보 활용이 중요할 때 쓰임


 Skip-gram: 한단어를 통해 나머지 문맥 학습 - 더 정확한 학습
 - 희귀 단어도 주변 단어를 예측하는 데 사용 가능
 - 단어의 다향한 의미 포착 가능
 - 학습 시간이 오래 걸림

 => 희귀 단어 처리와 다양한 의미 포착을 할 때 사용

```python
print("FastText CBOW 유사도:", transform(model.get_nearest_neighbors(word_to_jamo('남동생'), k=10)))
print("FastText Skip Gram 유사도:",  transform(model_sg.get_nearest_neighbors(word_to_jamo('남동생'), k=10)))
```

**방처럼 짧은 단어에서 Skip-gram이 더 좋은 이유**:

1. 단어별 정보 보존 => 중심 단어 하나씩에 대해 주변 단어를 학습하므로, 고유한 의미 표현을 더 잘 보존 ! 다양한 맥락에서 사용될 수 있는 단어는 skip-gram이 맥락별 정보를 효과적으로 학습!
2. 희석 문제 방지 => CBOW의 경우 주변 단어의 평균을 사용하기 때문에 중심정보에 대한 정보가 희석될 수 있다! 짧은 단어는 종종 더 많은 문맥에 걸쳐 사용됨
3. 위에서 예상한 것처럼 희귀단어와도 연관이 있는데, 짧은 단어는 빈도가 높고 다양한 문맥에 등장할 가능성이 크다. 근데 skip gram은 희귀단어 뿐 아니라 고빈도 단어에 대해서도 개별적 학습이 가능!
