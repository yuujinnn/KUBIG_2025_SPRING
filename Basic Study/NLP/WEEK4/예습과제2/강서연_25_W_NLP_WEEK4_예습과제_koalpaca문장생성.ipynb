```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
```

peft는 parameter efficient fine-tuning의 약자로, 큰 모델을 코랩에서 돌릴 수 있도록 쪼개서 만든 것입니다.

오래 걸립니다!

```python
from peft import PeftModel, PeftConfig
```

```python
peft_model_id = "beomi/qlora-koalpaca-polyglot-12.8b-50step"
config = PeftConfig.from_pretrained(peft_model_id)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
model = AutoModelForCausalLM.from_pretrained(config.base_model_name_or_path, quantization_config=bnb_config, device_map={"":0})
model = PeftModel.from_pretrained(model, peft_model_id)
tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)

model.eval()
```

next token generation

```python
prompt = '인간처럼 생각하고, 행동하는 \'지능\'을 통해 인류가 이제까지 풀지 못했던'
with torch.no_grad():
  tokens = tokenizer.encode(prompt, return_tensors='pt').to(device='cuda', non_blocking=True)
  gen_tokens = model.generate(tokens, do_sample=True, temperature=0.8, max_length=64)
  generated = tokenizer.batch_decode(gen_tokens)[0]

print(generated)
```

 QA

```python
def gen(x):
    q = f"### 질문: {x}\n\n### 답변:"
    # print(q)
    gened = model.generate(
        **tokenizer(
            q,
            return_tensors='pt',
            return_token_type_ids=False
        ).to('cuda'),
        max_new_tokens=50,
        early_stopping=True,
        do_sample=True,
        eos_token_id=2,
    )
    print(tokenizer.decode(gened[0]))
```

```python
gen('건강하게 살기 위한 세 가지 방법은?')
```

## top_k

```python
prompt = '인간처럼 생각하고, 행동하는 \'지능\'을 통해 인류가 이제까지 풀지 못했던'
with torch.no_grad():
  tokens = tokenizer.encode(prompt, return_tensors='pt').to(device='cuda', non_blocking=True)
  gen_tokens = model.generate(tokens, do_sample=True, temperature=0.8, max_length=100, top_k=10)
  generated = tokenizer.batch_decode(gen_tokens)[0]

print(generated)
```

```python
prompt = '인간처럼 생각하고, 행동하는 \'지능\'을 통해 인류가 이제까지 풀지 못했던'
with torch.no_grad():
  tokens = tokenizer.encode(prompt, return_tensors='pt').to(device='cuda', non_blocking=True)
  gen_tokens = model.generate(tokens, do_sample=True, temperature=0.8, max_length=100, top_k=100)
  generated = tokenizer.batch_decode(gen_tokens)[0]

print(generated)
```

- top_k가 낮을수록 일관성 및 예측 가능성을 우선시하며, 높을수혹 창의적이고 다양한 출력을 선호한다.
- top_k=50: 가장 포괄적이고 문맥적으로 관련성이 높은 텍스트를 생성한다. 프롬프트를 이어받는 내용임을 살펴볼 수 있다.
- top_k=10: 가장 짧은 출력을 생성했으며, 주제에 대한 얕은 답변을 생성했다. 이는 top_k가 낮을수록 모델의 생성이 가장 확률이 높은 선택지로 제한하기 때문에 이런 결과가 나타난다.
- top_k=100: 문맥에서 다소 벗어나고, 확률적으로 낮은 답변을 생성했다. 인공지능에서 벗어나 포스트 휴먼 리더가 되려는 기업에 대한 논의로 넘어가 관련 없는 출력을 주기도 함을 볼 수 있다.

## temperature

```python
prompt = '인간처럼 생각하고, 행동하는 \'지능\'을 통해 인류가 이제까지 풀지 못했던'
with torch.no_grad():
  tokens = tokenizer.encode(prompt, return_tensors='pt').to(device='cuda', non_blocking=True)
  gen_tokens = model.generate(tokens, do_sample=True, temperature=0.1, max_length=100, top_k=50)
  generated = tokenizer.batch_decode(gen_tokens)[0]

print(generated)
```

```python
prompt = '인간처럼 생각하고, 행동하는 \'지능\'을 통해 인류가 이제까지 풀지 못했던'
with torch.no_grad():
  tokens = tokenizer.encode(prompt, return_tensors='pt').to(device='cuda', non_blocking=True)
  gen_tokens = model.generate(tokens, do_sample=True, temperature=1.0, max_length=100, top_k=50)
  generated = tokenizer.batch_decode(gen_tokens)[0]

print(generated)
```

- temperature 매개변수 역시 top_k 변수와 비슷하게 작용한다. 0에 가까울수록 예측 가능하고 반복적인 텍스트를 생성하며, 1에 가까워질수록 다양하고 창의적인 답변을 생성한다.
- temperature=0.8:프롬프트 입력 내용에 따라 포괄적이고 문맥적으로 관련성이 높은 응답을 생성했다.
- temperature=0.1: top_k=10과 비슷하게 가장 짧고 단순한 출력을 생성한다. 앞선 결과와 비슷하게 모델의 선택지를 가장 확률이 높은 토큰으로 제한하여 창의성보다는 안정성과 예측 가능성을 우선시하는 경향을 보인다.
- tempertature=1.0: 문맥적으로 관련성이 없는 부분으로 답변을 이어나가는 것을 살펴볼 수 있다. '지능'에서 책의내용으로 넘어가는 등 내용이 전환되었음을 살펴볼 수 있다.
